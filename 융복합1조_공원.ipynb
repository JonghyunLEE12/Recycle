```python
df2=pd.read_csv('./서울시 주요 공원현황.csv',encoding='cp949')
```

```python
df2
```

```python
df2=df2[df2['위치'].str.contains('공원')]
```

```python
#a=df1.loc[:,'라벨명']
a=df2.loc[:,'공원명']
b=df2.loc[:,'X좌표(WGS84)']
c=df2.loc[:,'Y좌표(WGS84)']
park2=pd.concat([a,b,c],axis=1)
#park2
park2=park2.dropna(axis=0)
park2
```

```python
import numpy as np
from pyproj import Proj, transform
```

```python
proj_1 = Proj(init='epsg:2097')
proj_2 = Proj(init='epsg:4326')
```

```python
park1=park.copy()
park1

x_list=[]
y_list=[]

for idx, row in park1.iterrows():
    x, y = row['Y좌표'], row['X좌표']
    x_, y_ = transform(proj_1, proj_2, x, y)
    x_list.append(x_)
    y_list.append(y_)
```

```python
park1['lon']=x_list
park1['lat']=y_list
park1
```
