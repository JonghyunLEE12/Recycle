```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

from bs4 import BeautifulSoup as bs
import time
import re

driver = webdriver.Chrome('C:/Temp/chromedriver')
driver.get('https://www.shutterstock.com/ko/search/%EC%86%8C%EC%A3%BC?image_type=photo')
```

```python
images=[]
```

```python

for i in range(1,30):
    time.sleep(2)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(1)
    img=driver.find_elements_by_css_selector('#content > div.s_f_ff6d0 > div > div> div > main > div > div > div > div > div > a > img')
    time.sleep(1)
    for doms in img:
        url = doms.get_attribute('src')
        time.sleep(0.5)
        images.append(url)
        images = list(filter(None,images))
    time.sleep(0.5)
    driver.find_element_by_xpath("//*[@id='content']/div[2]/div/div[2]/div/main/div/div/div/div/a").send_keys(Keys.ENTER)

len(images)
```
